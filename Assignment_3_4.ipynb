```python
################################

# Student: Justin Hudson
# Instructor: Kaleab Gorfu
# CS 121: Python for DS and ML
# Date: 05/04/2026

################################

import random

def valid(number_string):
    if number_string.isdecimal() and int(number_string) >= 1 and int(number_string) <= 10:
        return True
    else:
        return False

def plural_of_try(number):
    if number == 1:
        return "try"
    else:
        return "tries"

print("\n")

counter = 0
random_number = random.randint(1,10)
while True:
    while True:
        user_input = input("Guess any number from 1 to 10: ")
        if valid(user_input):
            break
        print("Invalid Entry. Please enter an integer between 1 and 10 inclusive.")
    
    user_guess = int(user_input)
        
    if user_guess == random_number:
        counter += 1
        print("\nYou got it!")
        print(f"And it only took {counter} {plural_of_try(counter)}!\n")
        break
    elif user_guess < random_number:
        counter += 1
        print("\nToo low!\n")
    else:
        counter += 1
        print("\nToo high!\n")
```
