```python
################################

# Student: Justin Hudson
# Instructor: Kaleab Gorfu
# CS 121: Python for DS and ML
# Date: 14/04/2026

################################

import re

# Function to validate phone numbers
def validate_phone(number):
    ## Remove parentheses, hyphens, spaces, and plus signs
    cleaned_number = re.sub(r'[\(\)\-\s\+]', '', number)
    ## Check for 7-15 digits
    if re.match("^[0-9]{7,15}$", cleaned_number):
        return True
    else:
        return False

def validate_email(email):
    if re.match("^\S+@\S+\.\S+$", email):
        return True
    else:
        return False

def validate_zip(zip_code):
    if re.match("^[0-9]{5}$", zip_code) or re.match("^[0-9]{5}-[0-9]{4}$",zip_code):
        return True
    else:
        return False

user_input = input("Please enter a phone number -> ")
if validate_phone(user_input):
    print("Valid phone number")
else:
    print("Invalid phone number")

user_input = input("Please enter an email address -> ")
if validate_email(user_input):
    print("Valid email address")
else:
    print("Invalid email address")

user_input = input("Please enter a zip code -> ")
if validate_zip(user_input):
    print("Valid zip code")
else:
    print("Invalid zip code")
```
